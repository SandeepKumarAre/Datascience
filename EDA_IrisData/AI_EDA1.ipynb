```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Dowloaded IRIS data from Kaggle.com... 
#  https://www.kaggle.com/arshid/iris-flower-dataset/downloads/iris-flower-dataset.zip/1 

iris = pd.read_csv('IRIS.csv')
iris.head()
```

```python
print(iris.columns)
```

```python
iris.plot(kind='scatter',x='sepal_length',y='sepal_width')
```

```python
sns.set_style('whitegrid');
sns.FacetGrid(iris,hue='species',height=4) \
    .map(plt.scatter,'sepal_length','sepal_width') \
    .add_legend()
```

```python
sns.pairplot(iris,hue='species',height=3)
```

```python
# from Above pair plot... we can observe PW vs PL are more easy to separate the flowers types...

# We can simple model with if else condtions 

# if PW <= 1 & PL <=2 then its is Setosa...

# if PW>= 1 && PW <=1.8 & if PL < = 5 && PL >= 3 then its a Versicolor...

# else... Virginica...

```

```python
import numpy as np
iris_setosa = iris.loc[iris["species"] == "Iris-setosa"];
iris_virginica = iris.loc[iris["species"] == "Iris-virginica"];
iris_versicolor = iris.loc[iris["species"] == "Iris-versicolor"];
#print(iris_setosa["petal_length"])
plt.plot(iris_setosa["petal_length"], np.zeros_like(iris_setosa['petal_length']), 'o')
plt.plot(iris_versicolor["petal_length"], np.zeros_like(iris_versicolor['petal_length']), 'o')
plt.plot(iris_virginica["petal_length"], np.zeros_like(iris_virginica['petal_length']), 'o')

plt.show()
```

```python
# Cumulative Distribution Analysis... (CDF)..

```

```python
counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, 
                                 density = True)
```

```python
counts
```

```python
counts, bin_edges = np.histogram(iris_setosa['petal_length'],bins=10,density=True)
#print(counts)
#print(bin_edges)

pdf = counts/sum(counts)
print(pdf)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf)
plt.plot(bin_edges[1:],pdf)

counts, bin_edges = np.histogram(iris_setosa['petal_length'],bins=20,density=True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);


plt.show()
```

```python
np.histogram(iris_setosa['petal_length'],bins=10,density=True)
```

```python
print('Mean:')
print('Setosa mean',np.mean(iris_setosa['petal_length']))

# Adding Outlier...
print('Outlier...',np.mean(np.append(iris_setosa['petal_length'],50)))
print('Versicolor',np.mean(iris_versicolor['petal_length']))
print('Verginica',np.mean(iris_virginica['petal_length']))

print('\n')

print('Varianece:')
print('Setosa mean',np.var(iris_setosa['petal_length']))
print('Versicolor',np.var(iris_versicolor['petal_length']))
print('Verginica',np.var(iris_virginica['petal_length']))

print('\n')

print('Std:')
print('Setosa STD',np.std(iris_setosa['petal_length']))
print('Versicolor STD',np.std(iris_versicolor['petal_length']))
print('Verginica STD',np.std(iris_virginica['petal_length']))
```

```python
# Median... Percentile.... Quntile... IQR... MAD... 
```

```python
print("\ Median")
print(np.median(iris_setosa['petal_length']))
print(np.median(iris_versicolor['petal_length']))
print(np.median(iris_virginica['petal_length']))
# with adding outlier...
print(np.median(np.append(iris_setosa['petal_length'],50)))



print('Percentile...')
print(np.percentile(iris_setosa['petal_length'],np.arange(0,100,25)))
print(np.percentile(iris_setosa['petal_length'],np.arange(0,100,75)))

print(np.percentile(iris_setosa['petal_length'],90))
print(np.percentile(iris_setosa['petal_length'],85))



```

```python
from statsmodels import robust

print('Median Absolute Deviation...')

print(robust.mad(iris_setosa['petal_length']))
print(robust.mad(iris_versicolor['petal_length']))
print(robust.mad(iris_virginica['petal_length']))
```
